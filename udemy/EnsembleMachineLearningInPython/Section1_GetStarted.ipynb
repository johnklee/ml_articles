## <font color='darkblue'>All Data is the Same</font>
* How should you practice?
* Many people assume "practice" constitutes using ML models on specific datasets: predicting price of a house/moving rating / etc.
* At this stage, you should be comfortable with the idea that "all data is the same"
* Doesn't matter what you're predicting
* Data is still an NxD matrix (samples x features)
* Still preprocess numerical/categorical columns regardless of their "meaning"
* Meaning of data is irrelevant - logistic regression doesn't care if you're predicting price of a house vs. movie rating.

### <font color='darkgreen'>What is good practice?</font>
* Implementation is good practice
* A lot of college courses just do theory and then exams, maybe try to understand behavior of model on a conceptual level.
* That's "ok" practice
* Best is to learn details of an algorithm and then implement it
* If you come across an implementation video, you may want to stop and try by yourself first.
* Best way to hone your skills.
* Using the algorithms on real dataset is useful, I assume you are already doing that
* E.g. at your job/lab/ just on something you're interested in.
* If not, find something you're interested in
* Primary: implementation
* Secondary: use on data
* We do both in this course
* You should try on other datasets as well.

### <font color='darkgreen'>Plug-and-Play</font>
* Especially bad for deep learning, not so bad for classic ML methods and ensemble methods
* Ensemble methods are your best bet for a plug-and-play solution
* But we already learned how to use SKLearn in my last course
* You wouldn't even need to take this course
* Your could learn it in 15 minutes, it wouldn't even require a course.
* This course is for people who want to know the difference between a decision tree and random forest
* API is the same - results are different because theory is different.
* At your real job, you will probably plug-and-play all the time
* But to be a good data scientist, this would be in addition to learning how the algorithms work
* To understand why ensembles are good for plug-and-play, you need theory form this course.
