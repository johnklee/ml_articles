## <font color='darkblue'>PDF and CDF</font>
([link](https://www.udemy.com/course/deep-learning-prerequisites-the-numpy-stack-in-python/learn/lecture/19643750#overview)) In this lecture we are going to look at how to calculate the PDF in the CDF under the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Now let's check the supported/well know probability distribution:

```python
x = np.linspace(-6, 6, 1000)
```

```python
fx = norm.pdf(x, loc=0, scale=1)
```

```python
# Probability distribution of Normal distribution
plt.plot(x, fx);
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
# Cumulative distribution function
# https://en.wikipedia.org/wiki/Cumulative_distribution_function
plt.plot(x, Fx);
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
```

```python
plt.plot(x, logfx);
```

## <font color='darkblue'>Convolution</font>
([link](https://www.udemy.com/course/deep-learning-prerequisites-the-numpy-stack-in-python/learn/lecture/19643754#overview)) In this lecture we are going to do convolution convolution is a very common operation in many fields such as deep learning signal processing a digital image processing and statistics.<br/>
<br/>
In this lecture will apply convolution to an image to create a blurring filter.

```python
from PIL import Image

im = Image.open('images/lena.png')
```

```python
gray = np.mean(im, axis=2)
```

```python
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
# https://en.wikipedia.org/wiki/Gaussian_filter
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='gray')
```

```python
from scipy.signal import convolve2d

out = convolve2d(gray, filt)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray');
```
